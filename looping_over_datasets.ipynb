# Looping Over Datasets

```python
for filename in ["./data/gapminder_gdp_asia.csv", "./data/gapminder_gdp_africa.csv"]:
    print(filename)
```

```python
for filename in ["./data/gapminder_gdp_asia.csv", "./data/gapminder_gdp_africa.csv"]:
    file = open(filename)
    print(file.readline())
    file.close()
```
